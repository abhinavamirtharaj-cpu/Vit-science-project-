```python
from textblob import TextBlob
import json

def analyze_chat_message(text: str) -> dict:
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    if polarity > 0.1:
        category = "ðŸŸ¢ Positive"
        color = "ðŸŸ¢"  # Colab HTML for color below
    elif polarity < -0.1:
        category = "ðŸ”´ Negative"
        color = "ðŸ”´"
    else:
        category = "ðŸŸ¡ Neutral"
        color = "ðŸŸ¡"

    result = {
        "input_text": text,
        "polarity_score": round(polarity, 2),
        "sentiment_category": category,
        "color_emoji": color
    }

    print(f"{color} {text} â†’ {category} (Polarity: {polarity:.2f})")
    return result

# Test with chat messages
chat_messages = [
    "I love this messaging app!",
    "Terrible service, full of bugs.",
    "It's okay, works fine."
]

analyses = [analyze_chat_message(msg) for msg in chat_messages]

print("\nFull JSON output:")
print(json.dumps(analyses, indent=2))
```
